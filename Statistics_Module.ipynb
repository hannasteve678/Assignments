# Statistics Module Assignment - Solutions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Dataset.csv')
df.head()
```

```python
print(df.shape)
print(df.columns)
```

```python
df['Px'] = df['Px_h'] * df['Px_w']
print(df['Px'].head())
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Px'], kde=True)
plt.show()
```

```python
plt.figure(figsize=(12,8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, cmap='coolwarm')
plt.show()
```

```python
sns.scatterplot(x='RAM', y='Price', data=df)
plt.show()
```
