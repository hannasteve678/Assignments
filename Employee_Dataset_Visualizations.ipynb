# Employee Dataset - Visualizations


## Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('whitegrid')
```

## Load Dataset

```python
df = pd.read_csv('/content/Employee Dataset.csv')
print(df.head())
```

## Dataset Information

```python
print(df.info())
print(df.describe())
```

## Correlation Matrix Heatmap

```python
plt.figure(figsize=(10,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```


Heatmap helps understand the relationship between numerical columns in the employee dataset.

## Histogram

```python
df.hist(figsize=(12,10))
plt.show()
```


Histograms show the distribution of numerical features present in the dataset.

## Boxplot

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df.select_dtypes(include=np.number))
plt.xticks(rotation=45)
plt.title('Boxplot of Numerical Columns')
plt.show()
```


Boxplot helps identify outliers in the dataset.

## Countplot

```python
cat_cols = df.select_dtypes(include='object').columns
print(cat_cols)
```

```python
# Replace column name if needed
sns.countplot(x=df[cat_cols[0]])
plt.xticks(rotation=45)
plt.title('Countplot')
plt.show()
```


Countplot shows the frequency of categories in the selected column.

## Pairplot

```python
sns.pairplot(df.select_dtypes(include=np.number))
plt.show()
```


Pairplot helps compare multiple numerical variables together.
