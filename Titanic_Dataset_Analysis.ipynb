## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('whitegrid')
```

## Load Dataset

```python
df = pd.read_excel('/content/Titanic Dataset.xlsx')
df.head()
```

## Dataset Information

```python
print(df.shape)
print(df.columns)
print(df.info())
```

## Missing Values

```python
print(df.isnull().sum())
```

## Statistical Summary

```python
df.describe()
```

## Correlation Heatmap

```python
plt.figure(figsize=(10,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Observation
Heatmap helps identify relationships between numerical variables in the Titanic dataset.

## Survival Countplot

```python
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

### Observation
The graph shows the number of passengers who survived and who did not survive.

## Age Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'].dropna(), kde=True)
plt.title('Age Distribution')
plt.show()
```

### Observation
Most passengers were between 20 and 40 years old.

## Fare Distribution

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Fare'])
plt.title('Fare Boxplot')
plt.show()
```

### Observation
Boxplot helps identify outliers in fare values.
